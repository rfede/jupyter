```python
def F(coeff, g = 0):
    n, c_0, c_1 = len(coeff) - 1, 0, 0
    for k in range(n + 1): 
        c_1 = (c_1 * g + coeff[n-k] * (n-k))
        c_0 = (c_0 + coeff[n-k] ) * g
    return g - c_0/c_1

def solution(guess, coeff):
    approx = F(coeff, guess)
    steps = 1
    while guess != approx != 0: # approximate equality (given floating points' truncation)
        guess = approx
        approx = F(coeff, guess)
        steps += 1
        if steps > 100000: # threeshold (floating points' truncation or zero abscense may result in infinite loops)
            return (None, steps)
    return guess, steps
```

```python
from random import *
from sympy import *
x = Symbol('x')

for i in range(101):
    n = randint(2,20)
    coeffs = [random()+randint(-11,11) for k in range(n+1)]
    expr = sum([coeffs[k] * x**k for k in range(n+1)])
    guess = random()+randint(-9999,9999)
    sol, steps = solution(guess, coeffs)
    print('\nStarting with guess = '+str(guess)+', found ', sol, 'after', steps, 'steps')
    print('for', expr, end = '\n\n')
```
