```python
from sympy import *
from IPython.display import display, Markdown, Math
init_session(use_latex=True)
from math import comb
```

```python
def Spread(n):
    s=Symbol('s')
    p=0
    for k in range(1,n+1):
        p += 2 * n * (-4)**(k-1) * binomial(n+k,2*k)/(n+k) * s**k
    return  p



def CompositeSpread(n):
    nfactors = factorint(n)
    p = Symbol('s')
    for b in sorted(nfactors, reverse=True):
        for i in range(nfactors[b]):
            p = p.subs(Symbol('s'),Spread(b))
    return p

n, k, s = symbols('n,k,s')

def SpreadPartialRoots(n):
    k = n//2
    s = Symbol('s')
    p=0
    osw = ((-1)**(n + 1)+1) # switch for odd
    esw = ((-1)**n+1) #  switch for even
    for i in range(k+1):
        p += (-4*s)**i * binomial(k+i,k-i) * \
                ((n* s*(k + i)*osw+i*(2* i+1)*esw)/(2*(2* i + 1)*s*(k + i)) )
    return  s * (1-s)**(((-1)**(n)+1)//2) * p**2

for n in range(2,31):
    spread = Spread(n)
    root = SpreadPartialRoots(n)
    display(
                Markdown(
                    '{}'.format(latex(Indexed('S',n), mode='inline'))+\
                    '= {}'.format(latex(root, mode='inline'))+\
                    ' = {}'.format(latex(spread, mode='inline'))),
                Markdown(
                    '{}'.format(latex('Identity ', mode='inline'))+\
                    '{}'.format(latex(spread.expand()==root.expand(), mode='inline'))),
                Markdown('{}'.format(latex('', mode='inline')))
                    
            )
```
