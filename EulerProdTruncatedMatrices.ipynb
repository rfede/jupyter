```python
#generating code

import numpy as np
from sympy import *
from IPython.display import display, Markdown, Math
init_session(use_latex=True)

def one_plus_q_to_the_n(n):
    arr = np.array([[0 for col in range(n+1)] for row in range(n+1)])
    arr[0, 0], arr[1, 0], arr[1, 1] = 1, 1, 1
    for row in range(2, n+1):
        for col in range(n+1):
            arr[row, col] = arr[row-1, col] if col<row else arr[row-1, col] + arr[row-1, col-row]
    return arr

def one_minus_q_to_the_n(n):
    arr = np.array([[0 for col in range(n+1)] for row in range(n+1)])
    arr[0, 0], arr[1, 0], arr[1, 1] = 1, 1, -1
    for row in range(2, n+1):
        for col in range(n+1):
            arr[row, col] = arr[row-1, col] if col<row else arr[row-1, col] - arr[row-1, col-row]
    return arr
```
