```python
# generating code
q = [[]]
for n in range(1,11):
    current = 'q'*n
    p = [q[k][::] for k in range(n)]+[[]]
    p[0].append([])
    p[1].append(current)
    for k in range(1, n):
        for prev in q[k]:
            p[k+1].append(prev+'.'+current)
    for k in range(n+1):
        try:
            print(str(n)+'C'+str(k)+':', ', '.join(['['+str(el)+']' if el not in q[k] else str(el) for el in p[k]]))
        except:
            print(str(n)+'C'+str(k)+':', ', '.join(['['+str(el)+']' for el in p[k]]))
    print('\n')
    for k in range(n+1):
        try:
            print(str(n)+'C'+str(k)+':', ', '.join(['['+str(el.count('q'))+']' if el not in q[k] else str(el.count('q')) for el in p[k]]))
        except:
            print(str(n)+'C'+str(k)+':', ', '.join(['['+str(el.count('q'))+']' for el in p[k]]))
    print('_____________________________________________________\n\n')
    q = p
```
