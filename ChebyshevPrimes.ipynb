```python
from sympy import *
from IPython.display import display, Markdown, Math
init_session(use_latex=True)
from math import comb
```

```python
def Chebyshev(n):
    k = n//2
    def odd(i):
        return Integer((-1)**(k-i)*(comb(k+i+1, k-i)*2**(2*i) * n)/(k+i+1)) *x**(2*i+1) 
    def even(i):
        return Integer((-1)**(k-i)*(comb(k+i, k-i)*2**(2*i-1) * n)/(k+i)) *x**(2*i)
    x = Symbol('x')
    p = 0 
    f = odd if n % 2 else even
    for i in range(k+1):
        term = f(i)
        p += term
        #display(term)
    return p


def ChebyFactor(n):
    nfactors = factorint(n)
    p = Symbol('x')
    for b in sorted(nfactors, reverse=True):
        for i in range(nfactors[b]):
            p = p.subs(Symbol('x'),Chebyshev(b))
    return p
```

```python
for n in range(1,30):
    cheby = ChebyFactor(n)
    display(
            Markdown(
                    '{}'.format(latex(Indexed('T',n), mode='inline'))+\
                    '= {}'.format(latex(cheby, mode='inline'))+\
                    ' = {}'.format(latex(cheby.expand(), mode='inline'))
            )
            )
```
