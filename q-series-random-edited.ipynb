```python
#generating code
#generating code
import numpy as np
from sympy import *
import random
from IPython.display import display, Markdown, Math
init_session(use_latex=True)
def harriot(size, inv = False):
    M = [[1]+[0 for col in range(size)]]
    sign = -1 if inv else 1
    for row in range(1,size+1):
        M.append([M[row-1][0]*sign]+[M[row-1][col-1]+sign * M[row-1][col] for col in range(1, size+1)])
    return Matrix(M)

def getRandomPoly(size, symbl):
    p = 1
    for k in range(1, size):
        if (random.random()<2/k):
            sign = 1 if random.random()>.5 else -1
            p += sign * random.choice(range(1,10)) * symbl**k
    return p + O(symbl**(size+1))

def getRandomTable_1(size):
    q = Symbol('q')
    p = getRandomPoly(size, q)
    display(
            Markdown(
                    'Computing for s = {}'.format(latex(p, mode='inline'))
            )
    )
    
    s = 1 + O(q**(size+1))
    M = [[0 for col in range(size+1)] for row in range(size+1)]
    M[0][0] = 1
    for row in range(1, size+1):
        s = expand(s*p)
        for power, coeff in enumerate(reversed(poly((s.removeO())).all_coeffs())):
            M[row][power] = coeff
    return Matrix(M)

def getRandomTable_2(size):
    q = Symbol('q')
    p = getRandomPoly(size, q)
    display(
            Markdown(
                    'Computing for s = {}'.format(latex(p, mode='inline'))
            )
    )
    
    s = 0 + p
    M = [[0 for col in range(size+1)] for row in range(size+1)]
    for row in range(size+1):
        for power, coeff in enumerate(reversed(poly((s.removeO())).all_coeffs())):
            M[row][power] = coeff
        s = expand(s*p)
    return Matrix(M)


def exampleTable(expression, size = 24):
    q = Symbol('q')
    p = sympify(expression) + O(q**(size+1))
    display(
            Markdown(
                    'Computing for s = {}'.format(latex(p, mode='inline'))
            )
    )
    
    s = 0 + p
    M = [[0 for col in range(size+1)] for row in range(size+1)]
    for row in range(size+1):
        for power, coeff in enumerate(reversed(poly((s.removeO())).all_coeffs())):
            M[row][power] = coeff
        s = expand(s*p)
    return Matrix(M)
    
def trialExample(expression, size = 24):
    pinv = harriot(size, inv=True)
    ptransp = harriot(size, inv=False).T

    display(
            Markdown(
                    '{} = {}'.format(latex(Symbol('P^{-1}'), mode='inline'), latex(pinv, mode='inline'))
            ),
            Markdown(
                    '\n\n{} = {}'.format(latex(Symbol('P^{T}'), mode='inline'), latex(ptransp, mode='inline'))
            )

    )

    for trial in range(1):
        display(
            Markdown(
                    '\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n____________________________\n'+\
                    rf"""Trial number {trial+1}:"""
            )
        )
        M = exampleTable(expression, size)
        pinv_M = pinv*M
        pinv_M_ptransp = pinv_M * ptransp
        display(
                Markdown(
                    '{} = {}'.format(latex(Symbol('M'), mode='inline'), latex(M, mode='inline'))
                ),
                Markdown(
                    rf"""$P^{-1} \times M $"""+\
                    ' = {}'.format(latex(pinv_M, mode='inline'))
                ),
                Markdown(
                    rf"""$(P^{-1} \times M)^{(-1)} $"""+\
                    ' = {}'.format(latex(pinv_M**(-1), mode='inline'))
                ),
                Markdown(
                    rf"""$P^{-1} \times M \times P^T $"""+\
                    ' = {}'.format(latex(pinv_M_ptransp, mode='inline'))
                ),
                Markdown(
                    rf"""$(P^{-1} \times M \times P^T )^{-1}$"""+\
                    ' = {}'.format(latex(pinv_M_ptransp**(-1), mode='inline'))
                ),

        )
    

        
def trial_1(size = 12):
    pinv = harriot(size, inv=True)
    ptransp = harriot(size, inv=False).T

    display(
            Markdown(
                    '{} = {}'.format(latex(Symbol('P^{-1}'), mode='inline'), latex(pinv, mode='inline'))
            ),
            Markdown(
                    '\n\n{} = {}'.format(latex(Symbol('P^{T}'), mode='inline'), latex(ptransp, mode='inline'))
            )

    )

    for trial in range(10):
        display(
            Markdown(
                    '\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n____________________________\n'+\
                    rf"""Trial number {trial+1}:"""
            )
        )
        M = getRandomTable_1(size)
        pinv_M = pinv*M
        pinv_M_ptransp = pinv_M * ptransp
        display(
                Markdown(
                    '{} = {}'.format(latex(Symbol('M'), mode='inline'), latex(M, mode='inline'))
                ),
                Markdown(
                    rf"""$P^{-1} \times M $"""+\
                    ' = {}'.format(latex(pinv_M, mode='inline'))
                ),
                Markdown(
                    rf"""$(P^{-1} \times M)^{(-1)} $"""+\
                    ' = {}'.format(latex(pinv_M**(-1), mode='inline'))
                ),
                Markdown(
                    rf"""$P^{-1} \times M \times P^T $"""+\
                    ' = {}'.format(latex(pinv_M_ptransp, mode='inline'))
                ),
                Markdown(
                    rf"""$(P^{-1} \times M \times P^T )^{-1}$"""+\
                    ' = {}'.format(latex(pinv_M_ptransp**(-1), mode='inline'))
                ),

        )
        
        
    
def trial_2(size = 12):
    pinv = harriot(size, inv=True)
    ptransp = harriot(size, inv=False).T

    display(
            Markdown(
                    '{} = {}'.format(latex(Symbol('P^{-1}'), mode='inline'), latex(pinv, mode='inline'))
            ),
            Markdown(
                    '\n\n{} = {}'.format(latex(Symbol('P^{T}'), mode='inline'), latex(ptransp, mode='inline'))
            )

    )

    for trial in range(5):
        display(
            Markdown(
                    '\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n____________________________\n'+\
                    rf"""Trial number {trial+1}:"""
            )
        )
        M = getRandomTable_2(size)
        pinv_M = pinv*M
        pinv_M_ptransp = pinv_M * ptransp
        display(
                Markdown(
                    '{} = {}'.format(latex(Symbol('M'), mode='inline'), latex(M, mode='inline'))
                ),
                Markdown(
                    rf"""$P^{-1} \times M $"""+\
                    ' = {}'.format(latex(pinv_M, mode='inline'))
                ),
                Markdown(
                    rf"""$(P^{-1} \times M)^{(-1)} $"""+\
                    ' = {}'.format(latex(pinv_M**(-1), mode='inline'))
                ),
                Markdown(
                    rf"""$P^{-1} \times M \times P^T $"""+\
                    ' = {}'.format(latex(pinv_M_ptransp, mode='inline'))
                ),
                Markdown(
                    rf"""$(P^{-1} \times M \times P^T )^{-1}$"""+\
                    ' = {}'.format(latex(pinv_M_ptransp**(-1), mode='inline'))
                ),

        )


    

```
