```python
from sympy import *
from IPython.display import display, Markdown, Math
init_session(use_latex=True)
```

```python
def Spread(n):
    s=Symbol('s')
    p=0
    for k in range(1,n+1):
        p += binomial(n+k,n-k) * (-4*s)**k / (n+k)
    return  (-n*p/2).expand()



def SpreadSquared(n):
    m = n//2
    s = Symbol('s')
    p=0
    sign = (-1)**n 
    swe = (sign+1)//2 #  switch for even
    for k in range(m+1):
        p += binomial(n-k,k) * (-4*s)**(m-k) * (sign * (m-k-s*n)+m-k+s*n) /(2* (n-k) ) # common factor 1/2 inside
    return  ((1-s)**(swe) * (p**2 / s)) # common factor 1/2^2 should go here
```

```python
for n in range(1,31):
    spread = Spread(n)
    root = SpreadSquared(n)
    display(
                Markdown(
                    '{}'.format(latex(Indexed('S',n), mode='inline'))+\
                    '= {}'.format(latex(root, mode='inline'))+\
                    ' = {}'.format(latex(spread, mode='inline'))),
                Markdown(
                    '{}'.format(latex(spread.expand()==root.expand(), mode='inline'))+\
                    '{}'.format(latex(' Identity', mode='inline'))),
                Markdown('{}'.format(latex('', mode='inline')))
                    
            )
```
