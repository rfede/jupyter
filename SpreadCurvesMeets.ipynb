```python
# S.Goh factorization

from sympy import *
from IPython.display import display, Markdown, Math
init_session(use_latex=True)
```

```python
def Spread(n):
    s=Symbol('s')
    p=0
    for k in range(1,n+1):
        p += binomial(n+k,n-k) * (-4*s)**k / (n+k)
    return  (-n*p/2).expand()
```

```python
from random import randint

for i in range(20):
    m, n = randint(1,50), randint(1,50)
    while m == n:
        n = randint(1,100)
    if m > n:
        m, n = n, m
    d = n - m
    a = n + m
    display(
                Markdown(
                    '{}'.format(latex(Indexed('S',n)-Indexed('S',m), mode='inline'))+\
                    '= {}'.format(latex(factor(Spread(n)-Spread(m)), mode='inline'))+\
                    ' = {}'.format(latex(sqrt(Indexed('S',d)*Indexed('S',a)), mode='inline'))+\
                    '={}'.format(latex(sqrt(factor(Spread(d))*factor(Spread(a))), mode='inline')))
                    
            )
    print('\n\n\n\n')
    
```
